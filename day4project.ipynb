```python
import pandas as pd

data = {
    "Customer":["Ali","Sara","Ahmed","Ayesha","Usman","Hina","Bilal","Zara","Hamza","Noor",
                "Asad","Maham","Daniyal","Iqra","Farhan"],
    
    "Branch":["Peshawar","Peshawar","Islamabad","Islamabad","Lahore",
              "Lahore","Peshawar","Islamabad","Lahore","Peshawar",
              "Islamabad","Lahore","Peshawar","Islamabad","Lahore"],
    
    "Account_Type":["Savings","Current","Savings","Current","Savings",
                    "Current","Savings","Savings","Current","Savings",
                    "Current","Savings","Current","Savings","Current"],
    
    "Balance":[120000,80000,150000,60000,200000,
               95000,175000,130000,70000,110000,
               90000,160000,85000,140000,100000]
}

df = pd.DataFrame(data)
print(df)

```

```python
total_customers_per_branch = df.groupby("Branch")["Customer"].count()
print("\ntotal customers per branch:", total_customers_per_branch)
```

```python
total_Balance_per_Branch = df.groupby("Branch")["Balance"].sum()
print("\ntotal balance per branch:", total_Balance_per_Branch)
```

```python
average_Balance_per_Branch = df.groupby("Branch")["Balance"].mean()
print("\naverage balance per branch:", average_Balance_per_Branch)
```

```python
savacc = df["Account_Type"].value_counts()

print("Savings Accounts:", savacc["Savings"])
print("Current Accounts:", savacc["Current"])
```

```python
savings_and_accounts_holders_per_branch = df.groupby("Branch")["Account_Type"].value_counts()
print("\nSavings and Current account holders per branch:\n", savings_and_accounts_holders_per_branch)
```

```python
aggregation = df["Balance"].agg(["min", "max", "mean","count"])
print("\nBalance Aggregation:\n", aggregation)
```
